```python
import re

def parse_responses(responses):
    expr = r'\b(yes|no|weak)\b'
    verdict_dict = {}
    for k, val in responses.items():
        val = val.split('\n')
        assert len(val) == 100, print(f'{k} has {len(val)} lines')
        verdicts = []
        for line in val:
            verdict = re.search(expr, line, flags=re.IGNORECASE)
            if verdict:
                verdicts.append(verdict.group(1).lower())
            
        assert len(verdicts) == 100, print(f'{k} has {len(verdicts)}')
        verdict_dict[k] = verdicts

    return verdict_dict

glm_verdicts = parse_responses(glm_responses)
print('done')
llm_verdicts = parse_responses(llm_responses)
print('done')
```

### plots

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.transforms import blended_transform_factory
from matplotlib.patches import Patch
from matplotlib.legend_handler import HandlerTuple

def plot_verdict_bars(glm_verdicts, llm_verdicts, order=None, gap=0.08):
    verdicts = ["yes", "maybe", "no"]
    colors = {"yes": "#2ecc71", "maybe": "#f5c16c", "no": "#e74c3c"}
    alias = {"yes": "yes", "weak": "maybe", "maybe": "maybe", "no": "no"}

    if order is None:
        order = list(dict.fromkeys(list(glm_verdicts.keys()) + list(llm_verdicts.keys())))

    # def counts(d, k):
    #     v = d.get(k, [])
    #     if isinstance(v, (dict, Counter)):
    #         return [int(v.get(t, 0)) for t in verdicts]
    #     c = Counter(v)
    #     return [int(c.get(t, 0)) for t in verdicts]

    def counts(d, k):
        v = d.get(k, [])
        if isinstance(v, (dict, Counter)):
            c = Counter()
            for key, cnt in v.items():
                c[alias.get(key, key)] += int(cnt)
        else:
            c = Counter(alias.get(x, x) for x in v)
        return [int(c.get(t, 0)) for t in verdicts]


    x = np.arange(len(order))
    w = 0.34
    max_total = max(max(sum(counts(glm_verdicts, k)), sum(counts(llm_verdicts, k))) for k in order)

    fig, ax = plt.subplots(figsize=(max(6, 1.6*len(order)), 6.8))

    for j, data in enumerate([glm_verdicts, llm_verdicts]):  # GLM left, LLM right
        pos = x + (j - 0.5) * (w + gap)
        bottom = np.zeros(len(order))
        for i, t in enumerate(verdicts):
            vals = np.array([counts(data, k)[i] for k in order])
            ax.bar(pos, vals, width=w, bottom=bottom, color=colors[t],
                   edgecolor="black", linewidth=1.1, label=t if j == 0 else None)
            bottom += vals

    ax.set_ylim(0, max_total * 1.12)
    ax.set_xticks(x, order)
    ax.legend(frameon=False, ncol=3)

    trans = blended_transform_factory(ax.transData, ax.transAxes)
    y_text = -0.10  # under tick labels
    for i, k in enumerate(order):
        g = counts(glm_verdicts, k)
        l = counts(llm_verdicts, k)
        ax.text(x[i], y_text, f"G {g[0]}/{g[1]}/{g[2]}\nL {l[0]}/{l[1]}/{l[2]}",
                ha="center", va="top", fontsize=10, transform=trans)

    fig.subplots_adjust(bottom=0.30)
    guide_y = -0.3  # closer to tick labels
    ax.text(0.5, guide_y, "Left bar = GLM, Right bar = LLM",
            ha="center", va="top", fontsize=12, transform=ax.transAxes)
    fig.subplots_adjust(bottom=0.34)  # add space if it clips

    #ax.set_title("Triple Validity Check: 100 triples per a keyword", pad=10, fontsize=14)
    ax.tick_params(axis="x", labelsize=11)
    ax.set_xlabel('Entities', labelpad=38, fontsize=14)
    ax.set_ylabel('Triples', fontsize=14)
    return fig, ax

fig, ax = plot_verdict_bars(glm_verdicts, llm_verdicts); plt.show()
```
